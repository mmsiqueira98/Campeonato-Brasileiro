```python
import pandas as pd
from unidecode import unidecode
import csv
```

#Limpeza dos Dados

```python
df = pd.read_csv('/content/dados_de_partidas2.csv')
```

```python
df.head(1)
```

```python
for coluna in df.columns:
    df[coluna] = df[coluna].replace({"'": "", "\[": "", "\]": ""}, regex=True)
    df[coluna] = df[coluna].apply(lambda x: unidecode(str(x)))
```

```python
df.head(1)
```

```python
df[['time_casa', 'time_visitante']] = df['times'].str.split(', ', expand=True)

df.head(1)
```

```python
df[['Posse_casa', 'Posse_visitante']] = df['posse_bola'].str.split(',', expand=True)
df.head(1)
```

```python
df = df.drop('times', axis=1)
df = df.drop('posse_bola', axis=1)
```

```python
df.head(1)
```

```python
new_order = ['id', 'champ', 'time_casa', 'time_visitante', 'resultados_casa', 'resultados_visitante', 'Posse_casa', 'Posse_visitante','defesas_casa', 'defesas_visitantes', 'falta_casa', 'falta_visitante', 'cartao_amarelo_casa', 'cartao_amarelo_visitante', 'cartao_vermelho_casa', 'cartao_vermelho_visitante', 'chute_fora_casa', 'chute_fora_visitante', 'escanteios_casa', 'escanteios_visitante' ]
df = df.reindex(columns=new_order)
```

```python
df.head(1)
```

```python
df.head(10)
```

```python
df['Posse_casa'] = df['Posse_casa'].str.replace('%', '')
df['Posse_visitante'] = df['Posse_visitante'].str.replace('%', '')
```

```python
df.head(1)
```

Com os dados devidamente tratados, vamos transformá-los em csv e deixar salvo no github.

```python
df.to_csv('dados_atualizados.csv', index=False)
```
